## SIMULATIONS

#### Constants

```python
# Setting the constant variables

eps = [0.1, 0.2, 0.3]
decay = [0.99, 0.95, 0.9]
runs = 10
horizon = 1+50
batch_size = 10000
sigma_name_1 = "σ=0.001"
sigma_name_2 = "σ=0.005"
sigma_name_3 = "σ=0.01"
eps_name_1 = "ε=0.1"
eps_name_2 = "ε=0.2"
eps_name_3 = "ε=0.3"
decay_name_1 = "λ=0.99"
decay_name_2 = "λ=0.95"
decay_name_3 = "λ=0.90"
```

### Environment 1; sigma=0.001

```python
# Setting the rewards (CTR/period) for the simulations in environment 1

mean_A = random.uniform(0.02,0.07)
mean_B = random.uniform(0.02,0.07)
mean_C = random.uniform(0.02,0.07)
mean_D = random.uniform(0.02,0.07)
mean_E = random.uniform(0.02,0.07)
mean_F = random.uniform(0.02,0.07)

means_1 = [mean_A, mean_B, mean_C, mean_D, mean_E, mean_F] # probability of a click (here means=rewards)
sigma_1 = 0.001
environment_1 = Env(rewards=means_1)
```
